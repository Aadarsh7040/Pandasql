# Filtering

```python
where keyword is used to filter the data
```

```python
# operators
1. or
2. and
3. in
4. not in
5. <
6. >
7. between
8. not between
```

```python
# Filter out the data for which Type is Small
df[df['Type']=='Small']
```

```python
# Filter out the data for which Type is Small
ps.sqldf("select * from df where Type='Small'")
```

```python
# Filter out data for which Type is Small or Sporty
df[(df['Type']=='Small') | (df['Type']=='Sporty')]
```

```python
# Filter out data for which Type is Small or Sporty
ps.sqldf('select * from df where Type="Small" or Type="Sporty"')
```

```python
# Filter out data for which Type is Small or Sporty (in operator)
ps.sqldf('select * from df where Type in ("Small","Sporty")')
```

```python
# Filter out data for which Type is Midsize and Origin is USA
df[(df['Type']=='Midsize') & (df['Origin']=='USA')]
```

```python
# Filter out data for which Type is Midsize and Origin is USA
ps.sqldf('select * from df where Type is "Midsize" and Origin is"USA"')
```

```python
# Filter out data for which Type is Midsize and AirBags is Driver only
```

```python
# Filter out data for which AirBags is other than Driver only
df[df['AirBags']!='Driver only']
```

```python
# Filter out data for which AirBags is other than Driver only
ps.sqldf('select * from df where AirBags not in ("Driver only")')
```

```python
# Filter out data for which Price is greater than 30
df[df['Price']>30]
```

```python
# Filter out data for which Price is greater than 30
ps.sqldf('select * from df where Price>30')
```

```python
# Filter out data for which Length is in range 150 to 200

df[(df['Length']>149) & (df['Length']<201)]
```

```python
# Filter out data for which Length is in range 150 to 200

ps.sqldf('select * from df where Length between 150 and 200')
```

```python
# Filter out data for which weight is not in range 2500 to 3000

df[(df['Weight']<2500) | (df['Weight']>3000)]
```

```python
# Filter out data for which weight is not in range 2500 to 3000

ps.sqldf('select * from df where Weight not between 2500 and 3000')
```

# Sorting

```python
# Sort the data by price in ascending order

df.sort_values('Price')
```

```python
# Sort the data by price in ascending order

ps.sqldf('select * from df order by Price')
```

```python
# Sort the data by price in descending order

df.sort_values('Price',ascending=False)
```

```python
# Sort the data by price in descending order

ps.sqldf('select * from df order by Price desc')
```

# Grouping and Aggregation

## 1.Column level Aggregate

```python
# Find average Price
df['Price'].mean()
```

```python
# Find average Price
ps.sqldf('select avg(Price) from df')
```

```python
# Find max value of Weight
df['Weight'].max()
```

```python
# Find max value of Weight
ps.sqldf('select max(Weight) from df')
```

```python
# Find min value of Length
df['Length'].min()
```

```python
# Find min value of Length
ps.sqldf('select min(Length) from df')
```

## 2.Group level Aggregate

```python
# Find out type wise average price
df.groupby('Type')['Price'].mean()
```

```python
# Find out type wise average price
ps.sqldf('select Type,avg(Price) from df group by Type')
```

```python
# Find out origin wise minimum weight
df.groupby('Origin')['Weight'].min()
```

```python
# Find out origin wise minimum weight
ps.sqldf('select Origin,min(Weight) from df group by Origin')
```
